```python
import transformers
```

```python
# Load the lightweight exon classifier
from transformers import AutoModel

exon_classifier = AutoModel.from_pretrained(
    "schmojo/evo2-exon-classifier",
    trust_remote_code=True   # pulls the two .py files
).to('mps')
```

```python
exon_classifier
```

```python
# loop over the layers and print their names and shapes
for name, param in exon_classifier.named_parameters():
    print(f"{name}: {param.shape}")
    
```
